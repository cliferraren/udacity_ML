### Exploring the Enron dataset (emails + finances);



    - loads up the dataset (pickled dict of dicts).

    The dataset has the form:
    
        *  enron_data["LASTNAME FIRSTNAME MIDDLEINITIAL"] = { features_dict }

       *  {features_dict} is a dictionary of features associated with that person.
       
    You should explore features_dict as part of the mini-project, but here's an example to get you started:
    

    enron_data["SKILLING JEFFREY K"]["bonus"] = 5600000

---

#### Libraries

```python
#Import Libbraries
import pickle
import pandas as pd
import numpy as np
```

#### Data Location and Load

```python
#Location of the pickled data
dir = "../final_project/final_project_dataset.pkl"
```

```python
#Load the pickled data to Dataframe
enron = pd.read_pickle(dir)
```

#### Dictionary data into Dataframe

```python
enron = pd.DataFrame.from_dict(enron, orient='index', dtype=float)
enron[:5]
```

#### Data Exploration

```python
enron.info()
```

```python
enron.poi.unique()
```

```python
#Count number of Person of Interrest(POI)
poi_count = len(enron.loc[enron['poi'] == True])
print("Number of Person of Interest: {}".format(poi_count))
```

```python
enron.describe()
```

```python
missing = pd.DataFrame({'Null Values':enron.isna().sum(), 'percent':enron.isna().mean()*100})
```

```python
missing.sort_values(by='Null Values', ascending=False)
```

#### Load the Name List Text File

```python
nfile = '../final_project/poi_names.txt'
```

```python
namesdf = pd.read_csv(nfile, delimiter=')',names=('poi', 'Name'),\
                      index_col='Name', skiprows=[0])
namesdf[:5]
```

```python
namesdf = namesdf.replace('[(]', '', regex=True)
namesdf.sort_values(by='Name', ascending=True)
```
