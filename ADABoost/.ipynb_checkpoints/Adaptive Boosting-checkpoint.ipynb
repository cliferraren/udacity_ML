```python
#Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
from time import time
```

```python
#Import make terrain data and visualization
from class_vis import prettyPicture
from prep_terrain_data import makeTerrainData
```

```python
### features_train and features_test are the features for the training
### and testing datasets, respectively
### labels_train and labels_test are the corresponding item labels
features_train, labels_train, features_test, labels_test = makeTerrainData()
```

```python
## Import module to create Decision Trees classifier
from sklearn import tree
```

#### min_samples_split = 2

```python
#### Create the model
clf = tree.DecisionTreeClassifier(min_samples_split=2)
```

```python
### fit the classifier on the training features and labels
t0=time()
clf.fit(features_train,labels_train)
print("Training time: ",round(time()-t0,3),"secs")
```

```python
### use the trained classifier to predict labels for the test features
t1=time()
pred=clf.predict(features_test)
print("Prediction time: ",round(time()-t1,3),"secs")
```

```python
prettyPicture(clf, features_test, labels_test)
plt.show()
```

```python
#Get accuracy for the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(pred, labels_test)
print(accuracy *100,"%")
```

#### min_samples_split = 50

```python
#### Create the model
clf50 = tree.DecisionTreeClassifier(min_samples_split=50)
```

```python
### fit the classifier on the training features and labels
t50=time()
clf50.fit(features_train,labels_train)
print("Training time: ",round(time()-t50,3),"secs")
```

```python
### use the trained classifier to predict labels for the test features
t51=time()
pred50=clf50.predict(features_test)
print("Prediction time: ",round(time()-t51,3),"secs")
```

```python
prettyPicture(clf50, features_test, labels_test)
plt.show()
```

```python
#Get accuracy for the model
accuracy50 = accuracy_score(pred50, labels_test)
print(accuracy50 *100,"%")
```
